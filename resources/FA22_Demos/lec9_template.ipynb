```python
import pandas as pd
import numpy as np
from sklearn import datasets

iris = datasets.load_iris()
iris = pd.DataFrame(iris.data[:,1:3], columns=iris.feature_names[1:3])
iris.head()
```

```python
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

# use StandardScaler to normalize features
iris_nm = StandardScaler().fit_transform(iris)
```

```python
# make a scatter plot
plt.scatter(iris_nm[:,0], iris_nm[:, 1])
```

```python
# create a dendrogram
clust = linkage(iris)
dendrogram(clust)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# use fcluster to visualize the clusters better
fclust = fcluster(clust, 3, criterion = "maxclust")
plt.scatter(iris[:, 0], iris[:, 1], c=fclust)
```

```python
from sklearn.cluster import KMeans
# fit KMeans object
kmeans = KMeans(n_clusters = 5)
kmeans.fit(iris)
```

```python
# plot kmeans cluster results
labels = kmeans.labels_
d0 = iris.iloc[np.where(labels == 0)]
d1 = iris.iloc[np.where(labels == 1)]
d2 = iris.iloc[np.where(labels == 2)]

plt.scatter(iris["sepal width (cm)"], iris["petal length (cm)"], c = labels)
```

```python
# you can use these functions
from matplotlib.patches import Ellipse

def draw_ellipse(position, covariance, ax=None, **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()

    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)
    
    # Draw ellipses
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))
```

```python
from sklearn.mixture import GaussianMixture

# create GMM object
gmm = GaussianMixture(n_components = 3)
gmm.fit(iris)
predictions = gmm.predict(iris)
```

```python
# plot ellipses
plt.scatter(iris['sepal width (cm)'], iris['petal length (cm)'], c=predictions, s=40, cmap='viridis')

w_factor = 0.2 / gmm.weights_.max()
for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
  draw_ellipse(pos, covar, alpha=w * w_factor)
```
